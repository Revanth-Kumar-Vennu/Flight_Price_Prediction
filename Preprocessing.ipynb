```python
import pandas as pd

# Read CSV file into a dataframe
df = pd.read_excel('D:/IDMP/Project/Data/Flight_Data.xlsx')

# Display the dataframe
print(df)
```

```python

to_minutes = lambda x: int(x.split('h')[0]) * 60 + int(x.split(' ')[-1].replace('m', '')) if 'm' in x and 'h' in x else int(x.split('h')[0]) * 60 if 'h' in x else int(x.split('m')[0]) if 'm' in x else None

# Apply the lambda function to the 'time' column and create a new 'minutes' column
df['minutes'] = df['Duration'].apply(to_minutes)

# Print the DataFrame with the new 'minutes' column
print(df)
```

```python
import matplotlib.pyplot as plt

df.plot.scatter(x='Price', y='minutes')

# Show the plot
plt.show()
```
